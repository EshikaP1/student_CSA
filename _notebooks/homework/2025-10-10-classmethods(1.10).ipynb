```java
public class UltimateBattle {

    // Class representing a Robot fighter
    static class Robot {
        String name;
        int power;
        int health;
        static double fightDuration = 0.0;

        // Constructor
        public Robot(String name, int power, int health) {
            this.name = name;
            this.power = power;
            this.health = health;
        }

        // Instance method: attack another robot
        void attack(Robot opponent) {
            System.out.println(this.name + " attacks " + opponent.name + " for " + this.power + " damage!");
            opponent.health -= this.power;
            if (opponent.health < 0) {
                opponent.health = 0;
            }
            fightDuration += 1.5;
        }

        // Instance method: print current status
        void printStatus() {
            System.out.println(name + " → Power: " + power + ", Health: " + health);
        }

        // Static method: determine which robot is stronger
        static int strongerFighter(Robot r1, Robot r2) {
            if (r1.power > r2.power) {
                return 1;
            } else if (r2.power > r1.power) {
                return 2;
            } else {
                return 0;
            }
        }

        // Static method: start the battle
        static void beginBattle(Robot r1, Robot r2) {
            System.out.println("\n=== Battle Begins ===");
            while (r1.health > 0 && r2.health > 0) {
                r1.attack(r2);
                if (r2.health > 0) {
                    r2.attack(r1);
                }
            }
            System.out.println("\n=== Battle Ends ===");
            System.out.println("Duration: " + fightDuration + " units of time");

            if (r1.health > 0) {
                System.out.println("Winner: " + r1.name);
            } else if (r2.health > 0) {
                System.out.println("Winner: " + r2.name);
            } else {
                System.out.println("It's a draw!");
            }
        }
    }

    // Main method to test the class
    public static void main(String[] args) {
        Robot arnold = new Robot("Arnold", 25, 100);
        Robot barnold = new Robot("barnold", 20, 110);

        arnold.printStatus();
        barnold.printStatus();

        int stronger = Robot.strongerFighter(arnold, barnold);
        if (stronger == 1) {
            System.out.println(arnold.name + " is the stronger fighter.");
        } else if (stronger == 2) {
            System.out.println(barnold.name + " is the stronger fighter.");
        } else {
            System.out.println("Both fighters are evenly matched.");
        }

        Robot.beginBattle(arnold, barnold);
    }
}
```

```java
UltimateBattle.main(null);
```
